```python
import numpy as np
```

```python
def ode_fun(x, t):
    kappa = 0.003
    gamma_c = 0.0001
    g = 0.06
    omega_c=9.99
    omega_q=9.85
    omega_d=10.0
    epsilon = 0.01
    
    k = kappa
    y = gamma_c
    d = omega_c - omega_d
    D = omega_q - omega_d
    eps = epsilon

    dx = np.zeros(5)
    dx[0] = -0.5*k*x[0] + d*x[1] + 0.5*g*x[3]
    dx[1] = -0.5*k*x[1] - d*x[0] - 0.5*g*x[2] - eps 
    dx[2] = -0.5*y*x[2] + D*x[3] - 2*g*x[1]*x[4]
    dx[3] = -0.5*y*x[3] - D*x[2] + 2*g*x[0]*x[4]
    dx[4] = -y*(x[4]+1) + 2*g*(-x[0]*x[3]+x[1]*x[2])
                               
    return dx
    
```

```python
x0 = np.array([0, 0, 0, 0, -1])
```

```python
t = np.linspace(0, 1000, 1000)
```

```python
from scipy.integrate import odeint
sol = odeint(ode_fun, x0, t)
```

```python
import matplotlib.pyplot as plt
plt.plot(t, sol[:, 0], 'b', label='theta(t)')
plt.plot(t, sol[:, 1], 'g', label='omega(t)')
plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.show()
```

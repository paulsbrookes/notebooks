```python
import numpy as np
displace = 0
```

```python
numbers = []
for i in range(10):
    for j in range(10):
        numbers.append(str(i) + str(j))
```

```python
drive_frequencies = np.linspace(10.5065, 10.5125, 21)
```

```python
drive_frequencies = np.linspace(10.5065, 10.5125, 21)
remove = np.array([10.5065, 10.5085, 10.5085, 10.5125])
for freq in remove:
    index = np.argwhere(drive_frequencies==freq)
    drive_frequencies = np.delete(drive_frequencies, index)
```

```python
drive_frequencies1 = np.linspace(9.95, 10.05, 21)
drive_frequencies2 = np.linspace(9.98, 10.02, 11)
drive_frequencies = np.concatenate([drive_frequencies1, drive_frequencies2])
drive_frequencies = np.array(sorted(set(drive_frequencies)))
```

```python
endtime = '10000'
transmon_levels = '2'
cavity_levels = '50'
snap_to_snap = '250'
bins = '50000'
snapshots_number='200'
kappa='3.00000000e-03'
gamma_c='1.00000000e-04'
g='0.06'
epsilon='0.01'
omega_c='9.99'
omega_q='9.85000e+00'
samplesize='1000'
photon_number='0'
chi='-9.7000000000e-02'
```

```python
folder = '10MHz_up'
theta='0.0+00'
phi='0.0000000000e+00'

for idx, freq in enumerate(drive_frequencies):
    file = open(numbers[idx+displace] + 'up' + '.cfg', 'wb')
    file.write('folder='+folder+'\n')
    file.write('endtime='+endtime+'\n')
    file.write('transmon_levels='+transmon_levels+'\n')
    file.write('cavity_levels='+cavity_levels+'\n')
    file.write('snap_to_snap='+snap_to_snap+'\n')
    file.write('bins='+bins+'\n')
    file.write('snapshots_number='+snapshots_number+'\n')
    file.write('kappa='+kappa+'\n')
    file.write('gamma_c='+gamma_c+'\n')
    file.write('g='+g+'\n')
    file.write('epsilon='+epsilon+'\n')
    file.write('omega_c='+omega_c+'\n')
    file.write('omega_q='+omega_q+'\n')
    file.write('omega_d='+str(freq)+'\n')
    file.write('theta='+theta+'\n')
    file.write('phi='+phi+'\n')
    file.write('samplesize='+samplesize+'\n')
    file.write('photon_number='+photon_number+'\n')
    file.write('chi='+chi+'\n')
    file.close() 
```

```python
folder = '10MHz_down'
theta='3.141592654+00'
phi='0.0000000000e+00'

for idx, freq in enumerate(drive_frequencies):
    file = open(numbers[idx+displace] + 'down' + '.cfg', 'wb')
    file.write('folder='+folder+'\n')
    file.write('endtime='+endtime+'\n')
    file.write('transmon_levels='+transmon_levels+'\n')
    file.write('cavity_levels='+cavity_levels+'\n')
    file.write('snap_to_snap='+snap_to_snap+'\n')
    file.write('bins='+bins+'\n')
    file.write('snapshots_number='+snapshots_number+'\n')
    file.write('kappa='+kappa+'\n')
    file.write('gamma_c='+gamma_c+'\n')
    file.write('g='+g+'\n')
    file.write('epsilon='+epsilon+'\n')
    file.write('omega_c='+omega_c+'\n')
    file.write('omega_q='+omega_q+'\n')
    file.write('omega_d='+str(freq)+'\n')
    file.write('theta='+theta+'\n')
    file.write('phi='+phi+'\n')
    file.write('samplesize='+samplesize+'\n')
    file.write('photon_number='+photon_number+'\n')
    file.write('chi='+chi+'\n')
    file.close() 
```
